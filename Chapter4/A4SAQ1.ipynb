```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Load the Iris dataset CSV (assume 'Iris.csv' is in the same folder)
df = pd.read_csv("C:/Users/Admin/Desktop/Ram_Datamininig/Chapter4/iris.csv")

# Display first few rows to understand the data
print(df.head())

# Drop the ID column if it exists
if 'Id' in df.columns:
    df = df.drop('Id', axis=1)

# Since Apriori requires categorical data, we will discretize the numerical columns
# We'll convert each numerical attribute into categories like 'low', 'medium', 'high'

def discretize(column):
    bins = 3  # low, medium, high
    labels = ['low', 'medium', 'high']
    return pd.cut(column, bins=bins, labels=labels)

for col in df.columns[:-1]:  # exclude species column
    df[col] = discretize(df[col])

# Convert dataset to one-hot encoded format for Apriori
df_onehot = pd.get_dummies(df)

# Apply Apriori algorithm to find frequent itemsets with min support 0.3
frequent_itemsets = apriori(df_onehot, min_support=0.3, use_colnames=True)

print("Frequent Itemsets:")
print(frequent_itemsets)

# Generate association rules with minimum confidence of 0.7
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
