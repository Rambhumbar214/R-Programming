```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import pandas as pd

# Load Iris data
iris = load_iris()
X = iris.data

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dendrogram
linked = linkage(X_scaled, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, labels=iris.target)
plt.title('Hierarchical Clustering Dendrogram - Iris')
plt.show()

# Agglomerative clustering (3 clusters)
agglo = AgglomerativeClustering(n_clusters=3)
clusters = agglo.fit_predict(X_scaled)

# Compare clusters with actual species
df = pd.DataFrame({'Cluster': clusters, 'Species': iris.target})
print(pd.crosstab(df['Cluster'], df['Species']))
```
